```python
import pandas as pd
df=pd.read_csv("/content/Advertising.csv")
```

```python
df=df.iloc[:,1:len(df)]
df.head()
```

```python
x = df.drop('sales',axis=1)
y = df[["sales"]]
```

```python
x.head()
```

```python
y.head() #bagımlı degisken
```

```python
#statsmodel ile model kurma
import statsmodels.api as sm
```

```python
lm=sm.OLS(y,x)
```

```python
model = lm.fit()
```

```python
model.summary()
#rkare (R-squared) bagımsız degiskenlerin bagımlı degiskeni açıklama yüzdeki
#Prob (F-statistic): 0,05ten küçükse model anlamlı
```

```python
##scikit learn ile model kurma
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
model = lm.fit(x,y)
```

```python
#sınama seti
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.20, random_state=99)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
lm = LinearRegression()
model = lm.fit(X_train,y_train)
```

```python
np.sqrt(mean_squared_error(y_train, model.predict(X_train)))
#eğitim hatası
#egitim hatası varsa test hatası da vardir
```

```python
np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
```

```python
#k-katlı cv
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(model, X_train,y_train,cv=10,scoring="neg_mean_squared_error")
#cv=10 yani 10 katlı, 10 farklı hata hesaplar
```

```python
#cross validation(cv) MSE değeri
np.mean(-cross_val_score(model, X_train,y_train,cv=10,scoring="neg_mean_squared_error"))
#ortalama hata
```

```python
#RMSE
np.sqrt(np.mean(-cross_val_score(model, X_train,y_train,cv=10,scoring="neg_mean_squared_error")))
```
